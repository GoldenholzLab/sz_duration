```python
import random
import matplotlib.pyplot as plt

def roll_and_sum():
    rolls = [random.randint(1, 6) for _ in range(4)]
    return sum(sorted(rolls)[1:])  # Sort, drop lowest, and sum

def simulate_and_plot(num_simulations=1000):
    results = [roll_and_sum() for _ in range(num_simulations)]
    
    plt.figure(figsize=(10, 6))
    plt.hist(results, bins=range(min(results), max(results) + 2, 1), edgecolor='black')
    plt.title(f'Histogram of {num_simulations} Dice Rolls (4d6 drop lowest)')
    plt.xlabel('Sum of 3 highest dice')
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)
    plt.show()

# Run the simulation and plot
simulate_and_plot(10000000)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
total_days = 180
cycle_length = 30
seizure_days = 2
window_size = 90

# Create the seizure pattern
pattern = np.zeros(cycle_length)
pattern[:seizure_days] = 1
full_pattern = np.tile(pattern, total_days // cycle_length + 1)[:total_days]

# Calculate moving average
moving_avg = np.convolve(full_pattern, np.ones(window_size), 'valid') / window_size

# Create x-axis for plotting
x_axis = np.arange(window_size, total_days + 1)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(full_pattern, label='Seizure Occurrence')
plt.plot(x_axis, moving_avg)
plt.title('90-Day Moving Average of Seizure Occurrence')
plt.xlabel('Day')
plt.ylabel('Moving Average')
plt.ylim(0, 0.15)  # Adjust y-axis limit for better visibility
plt.grid(True)
plt.show()

# Print some values for verification
print(f"First 30 days of seizure pattern: {full_pattern[:30]}")
print(f"Moving average (first 5 values): {moving_avg[:5]}")
print(f"Moving average (last 5 values): {moving_avg[-5:]}")
```

```python
import random
import matplotlib.pyplot as plt

def roll_3d6():
    return sum(random.randint(1, 6) for _ in range(3))

def simulate_and_plot(num_simulations=10000):
    results = [roll_3d6() for _ in range(num_simulations)]
    
    plt.figure(figsize=(12, 6))
    plt.hist(results, bins=range(min(results), max(results) + 2, 1), edgecolor='black', align='left')
    plt.title(f'Histogram of {num_simulations} 3d6 Rolls')
    plt.xlabel('Sum of 3 dice')
    plt.ylabel('Frequency')
    plt.xticks(range(min(results), max(results) + 1))
    plt.grid(True, alpha=0.3)
    plt.show()

# Run the simulation and plot
simulate_and_plot(6)
```
